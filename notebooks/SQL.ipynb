```python
    
for index, row in clientes.iterrows():
    try:
        query = f"""
        INSERT INTO clientes (opp_id, name, email, city) 
        VALUES ('{row["Opportunity ID"]}', '{row["Name"]}', '{row["Email"]}', '{row["City"]}');
        """
        print(query)
        engine.execute(query)
    except:
        print(row["Opportunity ID"])
```

```python
for index, row in programa.iterrows():
    try:
        query = f"""
        INSERT INTO programa (opp_id, interest, format, start_date, campus) 
        VALUES ('{row["Opportunity ID"]}', '{row["Interest"]}', '{row["Format"]}', '{row["Start Date"]}');
        """
        print(query)
        engine.execute(query)
    except:
        print(row["Opportunity ID"])

```
