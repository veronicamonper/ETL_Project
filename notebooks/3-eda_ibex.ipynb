```python
import pandas as pd
```

```python
# 

df = pd.read_csv('../data/dataset_IBEX.csv', index_col = 0, sep=',')
df.head()
```

```python
# revisamos si hay nulos

df.isnull().sum()
```

```python
df.reset_index(inplace=True)

df.head()
```

```python
# cambiamos nombres de las columnas

df.rename(columns={'Fecha': 'Date'}, inplace=True)
df.rename(columns={'Último': 'Closure'}, inplace=True)
df.rename(columns={'Apertura': 'Opening'}, inplace=True)
df.rename(columns={'Máximo': 'Max'}, inplace=True)
df.rename(columns={'Mínimo': 'Min'}, inplace=True)
df.rename(columns={'Vol.': 'Euros'}, inplace=True)
df.rename(columns={'Semana': 'Week'}, inplace=True)
df.rename(columns={'Año': 'Year'}, inplace=True)


df.head()
```

```python
# cambiamos el formato de la fecha
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
# Quitamos el carácter "M y B" de los números en la columna 'Euros'
df['Euros'] = df['Euros'].str.replace('[MB]', '', regex=True)
df.head()

```

```python
df['% var.'] = df['% var.'].str.replace('%', '')
df.head()
```

```python
def format_numerical_value(value):
    # Quitamos el punto que separa los miles
    value = value.replace('.', '')
    # Sustituimos la coma por un punto
    value = value.replace(',', '.')
    return value

# Seleccionamos las columnas en las que queremos aplicar la transformación
columnas_a_transformar = ['Closure', 'Opening', 'Max', 'Min', 'Euros', '% var.']

# Aplicamos la función a las columnas seleccionadas del DataFrame
df[columnas_a_transformar] = df[columnas_a_transformar].applymap(format_numerical_value)

df.head()
```

```python
df['Week'] = df['Date'].dt.week
df.head()
```

```python
# Creamos una nueva columna 'Year' y extraemos el año de la columna 'Date'
df['Year'] = df['Date'].dt.year

df.head()
```

```python
df_limpio = df
df_limpio.head()
```

```python
df_limpio.to_csv('df_ibex.csv', index=False)
```
