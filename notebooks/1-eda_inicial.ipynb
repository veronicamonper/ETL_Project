# Librerias

```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/dataset_opps.csv", encoding = 'utf-8', index_col = 0, sep=';')
df1.head()      
```

```python
df1.columns
```

```python
df1.isnull().sum()
```

```python
df1.rename(columns={'Person Account: City Before Course': 'City'}, inplace=True)
df1.rename(columns={'Person Account: Email': 'Email'}, inplace=True)
df1.rename(columns={'Account Name': 'Name'}, inplace=True)
df1.rename(columns={'Bootcamp Start Date': 'Start Date'}, inplace=True)
df1.rename(columns={'Bootcamp Format': 'Format'}, inplace=True)
df1.rename(columns={'Campus extended': 'Campus'}, inplace=True)

df1.head()
```

```python
df1['City'].fillna('Other', inplace=True)
df1.head()
```

```python
df1.shape
```

```python


df1 = df1.drop_duplicates(subset='Email')
df1.tail()
```

```python
df_sin_duplicados = df1
df_sin_duplicados.head()
```

```python
#date time cambiar formato 
df_sin_duplicados['Start Date'] = pd.to_datetime(df_sin_duplicados['Start Date'])
df_sin_duplicados['Created Date'] = pd.to_datetime(df_sin_duplicados['Created Date'])
df_sin_duplicados['Close Date'] = pd.to_datetime(df_sin_duplicados['Close Date'])


df_sin_duplicados.tail()
```

```python
df_limpio = df_sin_duplicados.copy()
df_limpio.head()
```

```python
df_limpio.columns
```

```python
df_limpio.reset_index(inplace=True)

df_limpio.head()
```

```python
nuevo_orden_columnas = ['Opportunity ID', 'Name', 'Email', 'City', 'Campus', 'Stage', 'Created Date', 'Close Date', 'Stage Before Lost', 'Start Date', 'Format', 'Interest']

```

```python
df_limpio = df_limpio[nuevo_orden_columnas]
df_limpio.head()
```

```python
df_limpio.to_csv('eda_inicial.csv', index=False)
```

```python
df_limpio.columns
```

```python

# Seleccionamos columnas específicas para crear el nuevo DataFrame Clientes
col_clientes = ['Opportunity ID', 'Name', 'Email', 'City']

# Creamos el nuevo DataFrame con las columnas específicas
df_clientes = df_limpio[col_clientes]

df_clientes.head()
```

```python
df_clientes.to_csv('df_clientes.csv', index=False)
```

```python
# Seleccionamos columnas específicas para el nuevo DataFrame Created Date
col_created_date = ['Opportunity ID', 'Created Date', 'Interest', 'Format', 'Start Date']

# Creamos el nuevo DataFrame con las columnas específicas
df_created = df_limpio[col_created_date]

df_created.head()
```

```python
# Creamos una nueva columna 'Year' y extraemos el año de la columna 'Created Date'
df_created['Year'] = df_created['Created Date'].dt.year

df_created.head()
```

```python
# Creamos una nueva columna con el número de semana
df_created['Week'] = df_created['Created Date'].dt.week
df_created.head()
```

```python
nuevo_orden_created = ['Opportunity ID', 'Created Date', 'Interest', 'Format', 'Start Date', 'Week', 'Year']

df_created = df_created[nuevo_orden_created]

df_created.head()
```

```python
df_created.to_csv('df_created.csv', index=False)
```

```python
# Seleccionamos columnas específicas para el nuevo DataFrame Close Date
col_close_date = ['Opportunity ID', 'Close Date', 'Stage Before Lost', 'Interest', 'Format', 'Start Date']

# Creamos el nuevo DataFrame con las columnas específicas
df_close = df_limpio[col_close_date]

df_close.head()
```

```python
# Creamos una nueva columna 'Year' y extraemos el año de la columna 'Close Date'
df_close['Year'] = df_close['Close Date'].dt.year

df_close.head()
```

```python
# Creamos una nueva columna con el número de semana
df_close['Week'] = df_close['Close Date'].dt.week
df_close.head()
```

```python
nuevo_orden_close = ['Opportunity ID', 'Close Date', 'Stage Before Lost', 'Interest', 'Format', 'Start Date', 'Week', 'Year']

df_close = df_close[nuevo_orden_close]

df_close.head()
```

```python
df_close.to_csv('df_close.csv', index=False)
```

```python
# Seleccionamos columnas específicas para el nuevo DataFrame Programa
col_programa = ['Opportunity ID', 'Interest', 'Format', 'Campus', 'Start Date']

# Creamos el nuevo DataFrame con las columnas específicas
df_programa = df_limpio[col_programa]

df_programa.head()
```

```python
df_programa.to_csv('df_programa.csv', index=False)
```
