```python
# primero importamos las librer√≠as que vamos a necesitar

import pandas as pd
import numpy as np

# estas son pa pintar
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_ibex = pd.read_csv('../data/df_ibex.csv', index_col = 0, sep=',')
```

```python
df_ibex.head()
```

```python
df_ibex.reset_index(inplace=True)

df_ibex.head()
```

```python
# juntamos las columnas week y year

df_ibex['week_year'] = df_ibex['Week'].astype(str) + '_' + df_ibex['Year'].astype(str)

df_ibex.head()
```

```python
# repetimos con el segundo csv de 'close'

df_close = pd.read_csv('../data/df_close.csv', index_col = 0, sep=',')

df_close.head()
```

```python
df_close.reset_index(inplace=True)

df_close.head()
```

```python
# juntamos las columnas week y year

df_close['week_year'] = df_close['Week'].astype(str) + '_' + df_close['Year'].astype(str)

df_close.head()
```

```python
# juntamos los dataframes

df_merged = df_close.merge(df_ibex, on='week_year')

df_merged.head()
```

```python
# eliminamos las columnas que no queremos

df_merged = df_merged.drop(['Close Date',  'Start Date', 'Date', 'Closure', 'Opening', 'Max', 'Min', 'Euros', 'Week_y', 'Year_y'], axis=1)

df_merged.head()
```

```python
# reordenamos las columnas

column_order = ['Opportunity ID', "Format", 'Stage Before Lost', 'Interest', '% var.', 'Week_x', 'Year_x', 'week_year']

df_merged = df_merged.reindex(columns=column_order)
df_merged.tail()
```

```python
# cambiamos el nombre de las columnas Week y Year
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df_merged = df_merged.rename(columns={'Week_x': 'Week', 'Year_x': 'Year'})
```

```python
df_merged.head()
```

```python
def sacar_bootcamp(col):
    return col[:2]
```

```python
df_merged["bootcamp"] = df_merged["Interest"].apply(sacar_bootcamp)
```

```python
df_merged.head()
```

```python
fig1 = sns.countplot(x = "bootcamp", data = df_merged, order= ['WD', 'DA', 'UX', 'CY'])

fig1

# podemos ver que el curso de WD es el pierde m√°s candidatos en el proceso
```

```python
fig2 = sns.countplot(x = "bootcamp", hue = "Stage Before Lost", data = df_merged, order= ['WD', 'DA', 'UX', 'CY']);

fig2

# en esta gr√°fica podemos ver por programa y en qu√© stage los candidatos dejan el proceso
```

```python
figu = sns.countplot(x = "Format", hue = "Stage Before Lost", data = df_merged);

# Personalizar el gr√°fico
plt.xlabel('Formato')
plt.ylabel('Conteo')
plt.title('Conteo de bajas por formato y etapa')
plt.legend(fontsize='small')

figu

# En esta gr√°fica quer√≠amos ver si hab√≠a diferencia seg√∫n formato
```

```python
# Sacamos la media de la variaci√≥n de la bolsa por curso y formato

df_merged.groupby("Interest")["% var."].mean().reset_index()
```

```python

# Filtramos para obtener solo las semanas del a√±o 2022
var_2022 = df_merged[df_merged['Year'] == 2022]
var_2022 = var_2022.sort_values('Week')


plt.plot(var_2022['Week'], var_2022['% var.'])

plt.xlabel('Semana')
plt.ylabel('% var.')
plt.title('Variaci√≥n Ibex 2022')
plt.gcf().autofmt_xdate()
plt.subplots_adjust(bottom=0.4)
plt.xticks(var_2022['Week'])


plt.show()
```
